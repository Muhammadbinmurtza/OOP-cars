```python
from typing import Any
car_name : str = input("what is your car name")
car_name_or = car_name.lower()
class Cars():
    def __init__(self,car_speed:str, car_model:int) -> None:
        self.car_speed : str = car_speed
        self.car_model : int = car_model
class Tesla(Cars): #inheritance
    def __init__(self, car_speed : Any, car_model : Any , charge_time : Any, autopilot_gen : Any, car_tyres : str ) -> None:
        super().__init__(car_speed, car_model)
        self.charge_time : str = charge_time
        self.autopilot_gen : int = autopilot_gen
        self.car_tyres : str = car_tyres
    
    def tesla_input(self) -> None:
        if car_name_or == "tesla":
            self.car_speed = input("Please write the Tesla speed: ")
            self.car_model = int(input("What is your Tesla model? "))
            self.charge_time = input("What is your Tesla charge time? ")
            self.autopilot_gen = int(input("What is the autopilot generation of your Tesla? "))
            self.car_tyres = input("which company tyres does your car has?")

        else:
            pass
    def tesla_print(self):
        if car_name_or == "tesla":
            print(f"the car name is tesla")
            print(f"the car speed  is : {self.car_speed}")
            print(f"the car model is : {self.car_model}")
            print(f"the charging time of tesla is : {self.charge_time}")
            if self.autopilot_gen <= 3 and self.autopilot_gen > 0:
                print(f"the autopilot gen is : {self.autopilot_gen}")
            else:
                print("invaid value written of autopilot gen")
            print(f"the tyres your car has are {self.car_tyres}")
        else:
            pass
if car_name_or == "tesla":
    tesla : Tesla = Tesla("",0,"",0,"")
    tesla.tesla_input()
    tesla.tesla_print()
else:
    pass

class Toyota(Cars):
    def __init__(self, car_speed: str, car_model: int, Type : str, TimeToReach100 : float) -> None:
        super().__init__(car_speed, car_model)
        self.car_type : str = Type
        self.TimeToReach100 : int = TimeToReach100
    
    def toyota_input(self):
        if car_name_or == "toyota":
            self.car_speed = input("Please write the Toyota speed: ")
            self.car_model = int(input("What is your Toyota model? "))
            self.car_type = input("Is your car prosmetic or Manual")
            self.TimeToReach100 = float(input("how much time your car takes to reach 0 to 100 , write in seconds"))
        else:
            pass
    
    def toyota_print(self):
        if car_name_or == "toyota":
            print("the car name is Toyota")
            print(f"the car model is : {self.car_model}")
            if self.car_type == "manual" or self.car_type == "prosmetic":
                print(f"the car type is : {self.car_type}")
            print(f"the time car takes to reach from 0 to 100 is : {self.TimeToReach100} seconds")
        else:
            pass

if car_name_or == "toyota":
    toyota1 : Toyota = Toyota("",0,"",0)
    toyota1.toyota_input()
    toyota1.toyota_print()

class Mercedees(Cars):
    def __init__(self, car_speed: str, car_model: int, Type : str, TimeToReach100 : float, engine : str ) -> None:
        super().__init__(car_speed, car_model)
        self.car_type : str = Type
        self.TimeToReach100 : int = TimeToReach100
        self.engine : str = engine

    def mercedees_input(self):
        if car_name_or == "mercedees":
            self.car_speed = input("Please write the Mercedees speed: ")
            self.car_model = int(input("What is your Mercedees model? "))
            self.car_type = input("Is your car prosmetic or Manual")
            self.TimeToReach100 = float(input("how much time your car takes to reach 0 to 100 , write in seconds"))
            self.engine : str = input("please specify the engine model of your car")
        else:
            pass

    def mercedees_print(self):
        if car_name_or == "mercedees":
            print("the car name is Mercedees")
            print(f"the car model is : {self.car_model}")
            if self.car_type == "manual" or self.car_type == "prosmetic":
                print(f"the car type is : {self.car_type}")
            else:
                pass
            print(f"the time car takes to reach from 0 to 100 is : {self.TimeToReach100} seconds")
            print(f"the engine type is : {self.engine}")
        else:
            pass
if car_name_or == "mercedees":
    mercedees1 : Mercedees = Mercedees("",0,"",0,"")
    mercedees1.mercedees_input()
    mercedees1.mercedees_print()

class Astin_martin(Cars):
    def __init__(self, car_speed: str, car_model: int, timeto100 : float, enginemodel: str ) -> None:
        super().__init__(car_speed, car_model)
        self.timeto100 : float = timeto100
        self.enginemodel : str = enginemodel

    def astinmartin_input(self):
        if car_name_or == "astinmartin":
            self.car_speed = input("Please write the Astin Martin speed: ")
            self.car_model = int(input("What is your Astin Martin model? "))
            self.timeto100 = float(input("how much time your car takes to reach 0 to 100 , write in seconds"))
            self.enginemodel : str = input("please specify the engine model of your car")
        else:
            pass
    def astin_martin_print(self):
        if car_name_or == "astinmartin":
            print("the car name is Astin Martin")
            print(f"the car model is : {self.car_model}")
            print(f"the time car takes to reach from 0 to 100 is : {self.timeto100} seconds")
            print(f"the engine type is : {self.enginemodel}")
        else:
            pass
    
if car_name_or == "astinmartin":
    astinmartin1 : Astin_martin = Astin_martin("",0,0,"")
    astinmartin1.astinmartin_input()
    astinmartin1.astin_martin_print()


```
